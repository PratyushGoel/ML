```python
from sklearn import datasets
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

X, Y = datasets.load_boston(return_X_y=True)
X=np.append(X, np.ones(len(X), dtype='int').reshape(-1, 1), axis=1)
x_train, x_test, y_train, y_test = train_test_split(X, Y)
```

```python
def cost(X, Y, M):
    cost = 0
    for i in range(len(X)):
        cost += (Y[i] - (M*X[i]).sum())**2
    return cost/len(X)
```

```python
def fit(X, Y):
    itrs = 600
    alfa = .000001
    M = np.ones(X.shape[1])
    
    for i in range(itrs):
        M = step_gradient(M, X, Y, alfa)
        print(i, cost(X, Y, M))
    return M
```

```python
def step_gradient(M, X, Y, alfa):
    slope = np.zeros(len(M))
    
    for i in range(len(X)):
        for j in range(len(M)):
            slope[j]+=(-2/len(X))*(Y[i]-((M*X[i]).sum()))*X[i][j]
    
    M = M - alfa*slope
    return M
```

```python
def score(Yt, Yp):
    u = ((Yt - Yp)**2).sum()
    v = ((Yt - Yt.mean())**2).sum()
    return 1-u/v
```

```python
def predict(X, M):
    y_pred = np.array(len(X))
    for i in range(len(X)):
        y_pred[i] = (M*X[i]).sum
    return y_pred
```
